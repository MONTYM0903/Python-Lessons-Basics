##  6. Explain the concept of generators in Python and how they are defined ?
ANS
Generator: A function that returns an iterator, using yield instead of return.

They produce items one at a time and don’t store the whole sequence in memory.

Example:
- def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

- for num in count_up_to(3):
    print(num)  # 1 2 3

## **1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?**

```python
def sum_even(numbers):
    total = 0
    for n in numbers:
        if n % 2 == 0:
            total += n
    return total
print(sum_even([1, 2, 3, 4, 5, 6]))
```

## **3.Implement a Python function that takes a list of integers and returns a new list containing the squares of each number?**


```python
def squares(numbers):
    result = []
    for n in numbers:
        result.append(n * n)
    return result
print(squares([1, 2, 3, 4]))
```

## **4. Write a Python function that checks if a given number is prime or not from 1 to 200?**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
for i in range(1, 201):
    if is_prime(i):
        print(i, "is prime")
```

## **5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified terms?**

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            next_num = self.a + self.b
            self.a = self.b
            self.b = next_num
            self.count += 1
            return self.a

# Example
fib = Fibonacci(7)
for num in fib:
    print(num)
```

## **6. Write a generator function in Python that yield the powers of 2 up to a given exponent?**

```python
def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i

# Example:
for n in powers_of_two(5):
    print(n)
```

```python
def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

# Example (if "test.txt" exists):
# for line in read_lines("test.txt"):
#     print(line)
```

## **8. Use a lambda function in python to sort a list of tuples based on the second element of each tuple?**

## 9. Write a python program that uses map() to convert a list tempratures temperatures from Celsius to Fahrenheit?

```python
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda t: (t * 9/5) + 32, celsius))
print(fahrenheit)
```

## **10. Create a python  that uses filter() to remove all the vowels from a given string?**

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda c: c not in vowels, text))

# Example
print(remove_vowels("Hello, World!"))
```

## **11. Imagine a accounting routine used in a book shop. It works on a list with sublists, which look like this:**
Order Number Book Title and Author Quantity Price per Item             34587 Learning Python, Mark Lutz 4 40.95
98762 Programming Python, Mark Lutz 5 56.80
77226 Head First Python, Paul Barry 3 32.95
88112 EinfuhDhrung in Python3, Bernd Klein 3 24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- Euro if the value of the order is smaller than 100,00 Euro.

Write a Python program using lambda and map.


```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "EinfuhDhrung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda order: (
        order[0],
        order[2]*order[3] + (10 if order[2]*order[3] < 100 else 0)
    ),
    orders
))

print(result)
```
